```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Make 3 random blobs in 2d space
X, y_true = make_blobs(n_samples=100, centers=3, n_features=2, cluster_std=0.5, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=30);
```

```python
# Fit to X and predict class values for X
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
# Get the kmeans cluster centers, note that the dimensions of the centers are (n_clusters, n_features)
centers = kmeans.cluster_centers_
```

```python
plt.suptitle("3 random groups of points - the cluster centers were found using the K-means algorithm")
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap='jet')
plt.scatter(centers[:, 0], centers[:, 1], c='yellow', s=300, alpha=0.5);

print(X.shape)
print(centers)
print("---")
```
